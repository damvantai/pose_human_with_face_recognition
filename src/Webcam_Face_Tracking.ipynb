```python
def compare_points(centerpoints, point):
    if not centerpoints:
        return
    print(centerpoints)
    print(type(centerpoints))
    distance_centerpoint = [distance.euclidean(centerpoints[i], point) for i in range(len(centerpoints))]  
    index_distance_min = np.argmin(distance_centerpoint)
    distance_min = np.min(distance_centerpoint)
    return index_distance_min, distance_min
```

```python
video_capture = WebcamVideoStream(src=0).start()

while True:
    frame = video_capture.read()
    frame = cv2.flip(frame, 1)
    
    rgb_frame = frame[:, :, ::-1]
    face_locations = face_recognition.face_locations(rgb_frame)
    
#     for (top, right, bottom, left) in face_locations:
#         x_point_new = (top + bottom) / 2.
#         y_point_new = (left + right) / 2.
#         point = np.asarray([x_point_new, y_point_new])
        
#         if not centerpoints:
#             centerpoints.append(point)
#             namefromcenterpoint.append(name)
        
#         index_point_min, distance_min = compare_points(centerpoints, point)
        
#         if distance_min < (bottom - top) / 4.:
#             name = namefromcenterpoint[index_point_min] + "don't compute"
#             cv2.putText(frame, name, (left+6, top-6), font, 1.0, (0, 0, 255), 1)
#             centerpoints[index_point_min] = point
            
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)
    
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        x_point_new = (top + bottom) / 2.
        y_point_new = (left + right) / 2.
        point = np.asarray([x_point_new, y_point_new])
        
        if centerpoints:
            index_point_min, distance_min = compare_points(centerpoints, point)
            if distance_min < (bottom - top) / 4.:
                name = namefromcenterpoint[index_point_min] + str(distance_min) + "/" + str((bottom - top) / 4.)
                cv2.putText(frame, name, (left+6, top-6), font, 1.0, (0, 0, 255), 1)
                centerpoints[index_point_min] = point
                cv2.rectangle(frame, (left, top), (right, bottom), (255, 0, 0), 3)
                print("dont' compute")
                print(distance_min)
            else:
                distance_face_encodings = face_recognition.face_distance(known_face_encodings, face_encoding)
                min_distance = np.min(distance_face_encodings)
                index_point_min = np.argmin(distance_face_encodings)
                if min_distance < 0.4:
                    name = known_face_names[index_point_min]
                    print(name)
                    cv2.rectangle(frame, (left, top), (right, bottom), (255, 0, 0), 3)
                    cv2.putText(frame, name, (left + 6, top - 6), font, 1.0, (0, 0, 255), 1)
                    centerpoints.append(point)
                    namefromcenterpoint.append(name)
                    print("compute")
        else:
            distance_face_encodings = face_recognition.face_distance(known_face_encodings, face_encoding)
            min_distance = np.min(distance_face_encodings)
            index_point_min = np.argmin(distance_face_encodings)
            if min_distance < 0.4:
                name = known_face_names[index_point_min]
                print(name)
                cv2.rectangle(frame, (left, top), (right, bottom), (255, 0, 0), 3)
                cv2.putText(frame, name, (left + 6, top - 6), font, 1.0, (0, 0, 255), 1)
                centerpoints.append(point)
                namefromcenterpoint.append(name)
                print("compute")
    
    cv2.imshow("Frame", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
