```python
import cv2
import dlib
```

```python
redColor = (0, 0, 255)
faceCascade = cv2.CascadeClassifier("another/haarcascade_frontalface_default.xml")
video_capture = cv2.VideoCapture("/mnt/24eb92b8-4fc1-4f32-b05c-deaa999ce6cf/Documents/datasets/face/camera_supervisor/1_02_H_082018120000.avi")
```

```python
trackingFace = 0
tracker = dlib.correlation_tracker()

while True:
    rc, frame = video_capture.read()
    
    if not trackingFace:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray, 1.3, 5)
        maxArea = 0
        x = 0
        y = 0
        w = 0
        h = 0
        
        for (_x, _y, _w, _h) in faces:
            if _w * _h > maxArea:
                x = int(_x)
                y = int(_y)
                w = int(_w)
                h = int(_h)
                maxArea = w * h
        
        if maxArea > 0:
            tracker.start_track(frame, dlib.rectangle(x -10,
                                                     y - 20,
                                                     x + w + 10,
                                                     y + h + 20))
            trackingFace = 1
            
    if trackingFace:
        trackingQuality = tracker.update(frame)
        
        print(trackingQuality)
        if trackingQuality >= 5.75:
            tracked_position = tracker.get_position()
            t_x = int(tracked_position.left())
            t_y = int(tracked_position.top())
            t_w = int(tracked_position.width())
            t_h = int(tracked_position.height())
            cv2.rectangle(frame, (t_x, t_y),
                         (t_x + t_w, t_y + t_h), redColor, 3)
        else:
            trackingFace = 0
    cv2.imshow("frame", frame)
    if cv2.waitKey(33) == ord('a'):
        print("pressed a")
        break
```
