```python
from skimage import io
import face_alignment # very 
import matplotlib.pyplot as plt
import cv2
```

```python
fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._3D, enable_cuda=False, flip_input=False)

input = io.imread('/home/damvantai/Documents/github/open_source/Multi-class-BBox-Label-Tool/Images/001/15.jpg')
```

```python
plt.imshow(input)
```

```python
preds = fa.detect_faces(input)
```

```python
preds
```

```python
preds = fa.get_landmarks(input)
```

```python
for (x, y) in preds[0]:
    cv2.circle(input, (x, y), 1, (255, 0, 0), 5)
plt.imshow(input)
```
